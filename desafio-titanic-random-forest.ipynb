## $\large{\color{RoyalBlue}{1.}}$ $\large{\color{CadetBlue}{\textbf{Carregando as Bibliotecas ðŸ“š}}}$

```python
import numpy as np 
import pandas as pd 
import os
```

## $\large{\color{RoyalBlue}{2.}}$ $\large{\color{CadetBlue}{\textbf{Lendo os dados ðŸ‘€}}}$

```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data.head()
```

```python
teste_data = pd.read_csv('/kaggle/input/titanic/test.csv')
teste_data.head()
```

## $\large{\color{RoyalBlue}{3.}}$ $\large{\color{CadetBlue}{\textbf{AnÃ¡lise ExploratÃ³ria dos Dados ðŸ”Ž}}}$

```python
mulher = train_data.loc[train_data.Sex == 'female']["Survived"]
taxa_mulher = sum(mulher)/len(mulher)

print("% de mulheres sobreviventes:", taxa_mulher)
print(sum(mulher), 'total')
print(len(mulher), 'comprimento')
```

```python
homem = train_data.loc[train_data.Sex == 'male']["Survived"]
taxa_homem = sum(homem)/len(homem)

print("% de homens sobreviventes:", taxa_homem)
```

## $\large{\color{RoyalBlue}{4.}}$ $\large{\color{CadetBlue}{\textbf{CriaÃ§Ã£o do Modelo ðŸ§ }}}$

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(teste_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': teste_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Progresso salvo!")
```
