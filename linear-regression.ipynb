### Montando um gráfico de regressão com dados aleatórios

Para este estudo de regressão linear os dados de análise serão gerados aleatoriamente, utilizando da função `make_regression` do **sklearn.datasets.**

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

#Gerando uma massa de dados
x,y = make_regression(n_samples=200, n_features=1, noise=20, random_state=1)

#mostrando no gráfico
plt.scatter(x,y)
plt.show()
```

O próximo código tem como objetivo a criação do modelo de regressão linear, os parâmetros padrões foram explicitamente citados aqui apenas como fins educacionais, como não foi feito nenhuma alteração nos parâmetros padrões, poderia deixar vazio que estaria subentendido

```python
from sklearn.linear_model import LinearRegression

#Crição do Modelo   
modelo = LinearRegression()

modelo.fit(x,y)
```

O modelo de regressão linear é entendido como a seguinte equação da reta:

>$y = {\color{Yellow}m}x+{\color{Green}b}$

Onde:
- ${\color{Yellow}m}$ é coeficiente angular *(o ângulo da reta)*;
- ${\color{Green}b}$ coeficiente linear *(o momento que a reta corta o **eixo y**)*.

Essas variavéis podem ser obtidas através do sklearn da seguinte forma:

```python
#Coeficiente linear (b)
b = modelo.intercept_
b
```

```python
#Coeficiente angular (m)
m = modelo.coef_
m
```

Uma vez que temos todos os valores das variáveis, podemos traçar no gráfico original a melhor reta que poderia representar este conjunto de dados

```python
#mostrando o resultado
import numpy as np
plt.scatter(x,y)
xreg= np.arange(-3,3,1)
plt.plot(xreg,m*xreg-b,color='red') #Gráfico de regressão y=mx+b
plt.show()
```

### Separando dados de teste e treino

Para obter resultados mais precisos é importante separar da amostra uma porcentagem de dados para servirem como testes, enquanto o restante é destino para treinar o modelo, essa divisão para este exemplo ficou sendo 30% para os dados de teste, o restando ficando 70% para treino.

```python
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y,test_size = 0.30)
```

Essa função `fit` nada mais é que o momento que realizaa distribuição dos dados entre os eixos x e y, neste caso estaremos 'fitando' os dados de treino e medindo o resultado disto em relação aos dados de testes, ou seja, o quão preciso a reta do nosso modelo está para os dados de testes, dos quais são dados inéditos para este modelo.

```python
modelo.fit(x_treino,y_treino)
resultado = modelo.score(x_teste,y_teste)
print(resultado)
```
